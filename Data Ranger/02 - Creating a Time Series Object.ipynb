### Handling Missing Values

````python
df_comp.isna()
````

````python
df_comp.isna().sum()
````

To handle the missing values, we'll use ```fillna ``` method as follows :
1. **Front Filling** : Assigns the value of the previous period.
2. **Back Filling** : Assigns the value of the next period.
3. **Assigning the same value** : Assigning the average to all the missing values within the time series 



````python
df_comp.spx = df_comp.spx.fillna(method="ffill")
````

````python
df_comp.isna().sum()
````

````python
df_comp.ftse = df_comp.ftse.fillna(method="bfill")
````

````python
df_comp.isna().sum()
````

````python
df_comp.dax = df_comp.dax.fillna(value=df_comp.dax.mean())
````

````python
df_comp.isna().sum()
````

We gonna split our data to $80\%$ Training examples and $20\%$ Test examples.

````python
size = int(len(df_comp)*0.8)
````

````python
df = df_comp.iloc[:size]
df_test = df_comp.iloc[size:]
````
